# OBJETIVOS GENERALES

El curso **Databricks Finance Lab - Analítica Financiera Agéntica** tiene como objetivos generales:

## 1. Dominio de Conceptos Financieros Fundamentales
Desarrollar una comprensión sólida de los principios fundamentales de matemática financiera, valoración de instrumentos y análisis de portafolios, integrándolos con herramientas modernas de analítica de datos.

## 2. Competencia en Analítica de Datos Financieros
Capacitar a los estudiantes en el uso de plataformas modernas de analítica (Databricks) para procesar, analizar y visualizar grandes volúmenes de datos financieros de manera eficiente y escalable.

## 3. Aplicación de Inteligencia Artificial en Finanzas
Introducir a los estudiantes en el uso de agentes de IA (Genie Code) como herramienta de asistencia para resolver problemas financieros complejos, automatizar análisis y generar insights accionables.

## 4. Contextualización Latinoamericana
Analizar casos prácticos y datos reales del mercado financiero argentino y latinoamericano, conectando la teoría con la realidad económica regional.

## 5. Preparación Profesional Integral
Formar profesionales con competencias técnicas y analíticas que les permitan destacarse en áreas de finanzas corporativas, banca, inversiones, fintechs y consultoría financiera en la era de la transformación digital.

# OBJETIVOS ESPECÍFICOS

Al finalizar el curso, los estudiantes serán capaces de:

## Competencias Técnicas - Matemática Financiera

1. **Calcular y aplicar** el valor del dinero en el tiempo utilizando fórmulas de valor presente y valor futuro en contextos de inversión y financiamiento.

2. **Convertir y comparar** diferentes tipos de tasas de interés (nominal, efectiva, equivalente, real) y aplicarlas en decisiones financieras.

3. **Analizar y resolver** problemas de anualidades, amortización de préstamos y sistemas de pago (francés y alemán).

4. **Valorar instrumentos financieros** como bonos y acciones utilizando modelos de descuento de flujos de caja.

5. **Calcular e interpretar** ratios financieros clave para evaluar la salud financiera de empresas.

6. **Medir y gestionar** el riesgo y retorno de activos individuales y portafolios de inversión.

## Competencias en Analítica de Datos

7. **Utilizar Python** para implementar funciones financieras, procesar datos y automatizar cálculos complejos.

8. **Escribir consultas SQL** para extraer, transformar y analizar datos financieros almacenados en bases de datos.

9. **Crear visualizaciones** efectivas (gráficos, tablas, dashboards) para comunicar resultados financieros.

10. **Trabajar con datos reales** de empresas argentinas y latinoamericanas utilizando APIs financieras (Yahoo Finance, datos de CNV).

## Competencias en IA Generativa

11. **Interactuar con Genie Code** mediante consultas en lenguaje natural para resolver problemas financieros.

12. **Validar y depurar** código generado por IA, asegurando precisión y consistencia.

13. **Optimizar flujos de trabajo** combinando programación manual con asistencia de IA para maximizar productividad.

## Competencias Profesionales

14. **Analizar casos reales** del mercado argentino y latinoamericano aplicando los conceptos teóricos del curso.

15. **Tomar decisiones financieras informadas** basadas en análisis cuantitativos rigurosos.

# CONTENIDO DE LOS MÓDULOS

## Módulo 01: Fundamentos de Matemática Financiera
**Duración**: 3 semanas (12 horas)

### Notebook 1.1 - Valor del Dinero en el Tiempo
* Concepto fundamental del valor temporal del dinero
* Valor Presente (VP) y Valor Futuro (VF)
* Fórmulas de capitalización simple y compuesta
* Aplicaciones prácticas en inversiones y ahorro
* **Referencia libro**: Capítulo 5, Sección 5.1-5.2 (pág. 123-125)

### Notebook 1.2 - Tasas de Interés y Conversiones
* Tasa Nominal Anual (TNA) vs Tasa Efectiva Anual (TEA)
* Tasas proporcionales y tasas equivalentes
* Conversión entre diferentes períodos de capitalización
* Tasa real vs tasa nominal (ajuste por inflación)
* Tasa de descuento comercial
* **Referencia libro**: Capítulo 5, Sección 5.3 (pág. 126-131)

### Notebook 1.3 - Anualidades y Amortización
* Anualidades ordinarias y anticipadas
* Valor actual y valor futuro de anualidades
* Perpetuidades simples y crecientes
* Sistema Francés de amortización (cuota constante)
* Sistema Alemán de amortización (amortización constante)
* Tablas de amortización y cálculo de cuotas
* **Referencia libro**: Capítulo 5, Sección 5.4 (pág. 132-143)

### Notebook 1.4 - Estados Financieros y Análisis
* Balance General: estructura y componentes
* Estado de Resultados: ingresos, costos y márgenes
* Estado de Flujo de Efectivo: operativo, inversión, financiamiento
* Conexión entre estados financieros
* **Referencia libro**: Capítulos 2-4 (pág. 33-120)

## Módulo 02: Valoración de Instrumentos Financieros
**Duración**: 2 semanas (8 horas)

### Notebook 2.1 - Valoración de Bonos
* Características de los bonos: cupones, vencimiento, valor nominal
* Precio de un bono: descuento de flujos de caja
* Relación inversa entre precio y tasa de interés
* Duración y convexidad
* Bonos cupón cero vs bonos con cupones
* Valoración de bonos argentinos (contexto local)
* **Referencia libro**: Capítulo 6, Sección 6.1-6.3 (pág. 147-165)

### Notebook 2.2 - Valoración de Acciones
* Modelo de descuento de dividendos (DDM)
* Modelo de Gordon (crecimiento constante)
* Modelos de múltiples etapas de crecimiento
* Ratio Precio/Utilidad (P/E) y múltiplos comparables
* Valoración relativa vs valoración absoluta
* Análisis de empresas argentinas (ADRs)
* **Referencia libro**: Capítulo 6, Sección 6.4-6.6 (pág. 166-185)

---

## Módulo 03: Análisis Financiero y Ratios
**Duración**: 2 semanas (8 horas)

### Notebook 3.1 - Ratios Financieros
* **Ratios de Liquidez**: Liquidez corriente, Prueba ácida, Capital de trabajo
* **Ratios de Solvencia**: Endeudamiento, Cobertura de intereses, Apalancamiento
* **Ratios de Rentabilidad**: ROA, ROE, Margen neto, Margen operativo
* **Ratios de Eficiencia**: Rotación de activos, Rotación de inventarios, Período de cobro
* **Ratios de Mercado**: P/E, P/B, Dividend Yield, EPS
* Análisis comparativo (benchmarking)
* Interpretación y aplicación práctica
* **Referencia libro**: Capítulos 3-4 (pág. 67-120)

## Módulo 04: Series Temporales y Gestión de Portafolios
**Duración**: 3 semanas (12 horas)

### Notebook 4.1 - Riesgo y Rentabilidad
* Retorno esperado y retorno histórico
* Medidas de riesgo: varianza, desviación estándar, VaR
* Distribución de retornos y análisis de normalidad
* Relación riesgo-retorno
* Coeficiente de Sharpe, Sortino y Treynor
* Beta y riesgo sistemático vs no sistemático
* **Referencia libro**: Capítulo 7, Sección 7.1-7.3 (pág. 189-215)

### Notebook 4.2 - Construcción de Portafolios
* Teoría moderna de portafolios (Markowitz)
* Diversificación y reducción de riesgo
* Frontera eficiente y portafolio óptimo
* Modelo CAPM (Capital Asset Pricing Model)
* Asignación de activos y rebalanceo
* Optimización con restricciones
* **Referencia libro**: Capítulo 7, Sección 7.4-7.6 (pág. 216-245)

---

## Módulo 05: Analítica Agéntica con IA
**Duración**: 5 semanas (20 horas)

### Notebook 5.1 - Introducción a Genie Code para Finanzas
* Qué es Genie Code y cómo funciona
* Ventajas de IA en análisis financiero
* Cómo formular consultas efectivas
* Mejores prácticas con agentes de IA
* Casos de uso en finanzas

### Notebook 5.2 - Análisis de Datos Financieros con IA
* Automatización de cálculos financieros
* Generación de visualizaciones con IA
* Debugging y validación de código
* Análisis exploratorio asistido por IA
* Optimización de flujos de trabajo

### Notebook 5.3 - Caso Integrador: Portfolio Inteligente
* Proyecto final integrador
* Construcción de un portafolio usando IA
* Análisis completo de riesgo-retorno
* Optimización y backtesting
* Presentación de resultados y recomendaciones
